假設方程式 f(x)=x^4-5x^3-6x^2+7x+5 其其微分方程式為 f'(x)=4x^3-15x^2-12x+7

```python
def f(x):
    return x**4 - 5*x**3- 6*x**2 + 7*x + 5
def df(x):
    return 4*x**3-15*x**2-12*x+7
```

```python
x=np.arange(-2,6,0.01)

plt.plot(x,f(x))
plt.ylim(-140,70)
plt.grid()
plt.xlabel("x-axis")
plt.ylabel("y-axis")

```

```python
plt.plot(x,f(x))
k=6
plt.scatter(k,f(k))
constant=0.001
for i in range(100):
    k=k-constant*df(k)
    plt.scatter(k,f(k))
```
