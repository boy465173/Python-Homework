```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib as mpl
```

```python
mpl.rc('font', family='Noto Sans CJK TC')
```

103年資料

```python
df = pd.read_csv('103_service area.csv')
```

```python
df.head()
```

```python
import locale
locale.setlocale(locale.LC_NUMERIC, '')
```

```python
df['1月'] = df['1月'].apply(locale.atof)
df['2月'] = df['2月'].apply(locale.atof)
df['3月'] = df['3月'].apply(locale.atof)
df['4月'] = df['4月'].apply(locale.atof)
df['5月'] = df['5月'].apply(locale.atof)
df['6月'] = df['6月'].apply(locale.atof)
df['7月'] = df['7月'].apply(locale.atof)
df['8月'] = df['8月'].apply(locale.atof)
df['9月'] = df['9月'].apply(locale.atof)
df['10月'] = df['10月'].apply(locale.atof)
df['11月'] = df['11月'].apply(locale.atof)
df['12月'] = df['12月'].apply(locale.atof)
```

```python
df['總計'] = df['1月'] + df['2月']+ df['3月']+ df['4月']+ df['5月']+ df['6月']+ df['7月']+ df['8月']+ df['9月']+ df['10月']+ df['11月']+ df['12月']
```

```python
df.head()
```

```python
df.drop(['1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df2 = df.sort_values(by = ['總計'], ascending=False)
```

```python
df2[:14].plot.bar(x='區名', y=['總計'])
```
